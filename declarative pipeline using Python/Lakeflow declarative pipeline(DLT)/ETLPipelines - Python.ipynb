```python
#decorators

def greet():
    print("hello!")
```

```python
greet()
```

```python
def my_decorator(func):
    def my_wrapper():
        print("before func execute me")
        func()
        print("after func execute me")
    return my_wrapper
```

```python
@decorator
def greet():
    print("hello!")
```

```python
greet()
```
