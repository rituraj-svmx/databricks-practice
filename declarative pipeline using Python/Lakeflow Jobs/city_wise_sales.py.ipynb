```python
from pyspark.sql.functions import col,sum

spark.sql("create schema if not exists lakeflow_job.gold")

orders_df = spark.table("lakeflow_job.silver.orders_dedup_silver")
customers_df = spark.table("lakeflow_job.silver.customers_enriched_silver")

total_sales_df = orders_df.join(customers_df,"customer_id", "inner").groupBy(customers_df.state,customers_df.city).agg(sum("total_amount").alias("total_sales"))
```

```python
total_sales_df.write.format("delta").mode("overwrite").saveAsTable("lakeflow_job.gold.total_sales")
```
