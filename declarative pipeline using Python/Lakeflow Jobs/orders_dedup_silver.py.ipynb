```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window

spark.sql("create schema if not exists lakeflow_job.silver")

orders_dedup = spark.table("lakeflow_job.silver.orders_cleaned_silver")

window = Window.partitionBy("order_id").orderBy(col("order_date").desc())

orders_with_rn = orders_dedup.withColumn("rn", row_number().over(window))

orders_with_rn.show()

```

```python
order_dedup = orders_with_rn.filter(col("rn")==1).drop("rn")
```

```python
order_dedup.write.mode("overwrite").saveAsTable("lakeflow_job.silver.orders_dedup_silver")
```
