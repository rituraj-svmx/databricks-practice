```python
from pyspark.sql.functions import col, sum , current_timestamp

dbutils.widgets.text("state", "", "state")

spark.sql("create schema if not exists lakeflow_job.gold")

state_param = dbutils.widgets.get("state")

print(f"processing for {state_param}")

total_sales_df = spark.table("lakeflow_job.gold.total_sales")

state_wise_sales_df = total_sales_df.filter(col("state") == state_param)

final_state_df = state_wise_sales_df.withColumn("large_city", col("total_sales") > 10000)
final_state_df.write.format("delta").mode("overwrite").saveAsTable(f"lakeflow_job.gold.state_wise_sales_{state_param}")

```
