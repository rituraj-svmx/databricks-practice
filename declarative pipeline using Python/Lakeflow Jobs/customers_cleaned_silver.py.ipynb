```python
from pyspark.sql.functions import col,current_timestamp,to_date

dbutils.widgets.text("catalog_name", "lakeflow_job", "catalog name")
catalog = dbutils.widgets.get("catalog_name")

spark.sql(f"create schema if not exists {catalog}.silver")

customers_df = spark.table("lakeflow_job.bronze.customers_bronze")
```

```python
customers_df.printSchema()
```

```python
customers_df.filter(col("EffectiveEndDate") =='-')
```

```python
customers_cleaned = (customers_df.filter(col("EffectiveEndDate")=="-").
 select(col("CustomerID").alias("customer_id"),
 col("CustomerName").alias("customer_name"),
 col("ContactNumber").alias("contact_number"),
 col("Email").alias("email"),
 col("Address").alias("city"),
 to_date(col("DateOfBirth")).alias("date_of_birth"),
 to_date(col("RegistrationDate")).alias("registration_date")
))
```

```python
customers_cleaned.write.format("delta").mode("overwrite").saveAsTable("lakeflow_job.silver.customers_cleaned_silver")
```
