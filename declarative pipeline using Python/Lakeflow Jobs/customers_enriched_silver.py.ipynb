```python
spark.sql("create schema if not exists lakeflow_job.silver")

customers_cleaned_df = spark.table("lakeflow_job.silver.customers_cleaned_silver")
```

```python
lookup_df = spark.read.format("json").option("header", "true").option("inferSchema", "true").load("/Volumes/lakeflow_job/default/ranjan_volume/inputdata/state_lookup/")
```

```python
customers_enriched_df = customers_cleaned_df.join(lookup_df,customers_cleaned_df.city == lookup_df.city, "left").drop(lookup_df.city).drop(lookup_df._corrupt_record)
```

```python
customers_enriched_df.write.format("delta").mode("overwrite").saveAsTable("lakeflow_job.silver.customers_enriched_silver")
```
