```python
def my_decorator(func):
    def my_wrapper():
        print("before func execute me ")
        func()
        print("after func execute me")
    return my_wrapper
```

```python
@my_decorator
def greet():
    print("Hello world")
```

```python
greet()
```

### create table from data frame using decorator

```python
def write_to_delta(func):
    def my_wrapper():
        df = func()
        table_name = func.__name__
        df.write.format("delta").saveAsTable("sales_catalog.default."+table_name)
    return my_wrapper
```

```python
@write_to_delta
def orders_bronze_latest():
    df = (spark.read.
          format("csv")
          .option("header", 'true')
          .option("inferSchema", "true")
          .load("/Volumes/sales_catalog/default/sales_volume/sales_data/orders"))
    return df
  
```

```python
orders_bronze_latest()
```

```python
%sql
select * from sales_catalog.default.orders_bronze_latest;
```
