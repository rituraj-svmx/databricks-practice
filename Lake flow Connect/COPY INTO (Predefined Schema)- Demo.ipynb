### Add additional column into a table

```python
%sql
drop table if exists orders;

create table orders(
  order_id int,
  order_date string,
  customer_id int,
  order_status string,
  _file_name string,
  _load_data timestamp
)
using delta
```

```python
%sql
copy into orders
from (select cast(order_id as int) as order_id, order_date, cast(customer_id as int) as customer_id, order_status
, _metadata.file_path as _file_name,
current_timestamp() as _load_data
from "/Volumes/workspace/default/lateflow_connect_volume/orders_data")
fileformat = CSV
FORMAT_OPTIONS(
    'header' = 'true'
)
```

```python
%sql 
desc extended orders;
```

```python
%sql
select * from orders;
```
