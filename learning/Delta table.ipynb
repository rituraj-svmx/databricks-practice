```python
data = [(101,"sku-1001", "wireless-mouse", "electronics", 2, 1002.2),
        (102,"sku-1002", "laptop", "electronics", 20, 100.2),
        (103,"sku-1003", "mobile", "electronics", 28, 300.2)]

schema = ["order_id", "product_id", "product_name", "category", "quantity", "price"]

df = spark.createDataFrame(data, schema)
```

```python
display(df)
```

```python
volume_path = '/Volumes/workspace/default/orders_volume/orders_data1'
df.write.format("delta").mode("overwrite").save(volume_path)
```
