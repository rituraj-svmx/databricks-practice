```python
data = [("Ranjan",101,200),("Golu",102,100),("Betu",103,300)]


df = spark.createDataFrame(data,["name","id","marks"])
display(df)
```

```python
dbutils.fs.mkdirs("/Volumes/workspace/default/orders_volume/studentdata")
```

```python
df.write.format("delta").mode("overwrite").save("/Volumes/workspace/default/orders_volume/studentdata")
display(spark.read.format("delta").load("/Volumes/workspace/default/orders_volume/studentdata"))
```
