## Read Delta Tables with Iceberg client
#### One Parquet file(data) , 1 delta metadata & 1 iceberg metadata

```python
%sql
select current_catalog();
```

```python
%sql
drop table if exists orders_managed;

create table orders_managed (
  order_id long,
  sku string,
  product_name string,
  category string,
  quantity long,
  price double
)
tblproperties ('delta.columnMapping.mode' = 'name',
'delta.enableIcebergCompatV2' = 'true',
'delta.universalFormat.enabledFormats' = 'iceberg'
)
```

```python
%sql
desc extended orders_managed;
```

```python
%sql
show tblproperties orders_managed;
    
```
